```r
# Identify categorical variables
categorical_vars <- names(df)[sapply(df, function(x) is.factor(x) || is.character(x))]
```

```r
# Name of the continuous variable
continuous_var <- "X..of.times.saw.a.doctor.in.past.year"

# Initialize an empty data frame to store ANOVA results and effect sizes
anova_results <- data.frame(
  Categorical_Variable = character(),
  F_Value = numeric(),
  P_Value = numeric(),
  Eta_Squared = numeric(),
  stringsAsFactors = FALSE
)

# Loop through each categorical variable and perform ANOVA
for (cat_var in categorical_vars) {
  # Ensure the categorical variable is a factor
  df[[cat_var]] <- factor(df[[cat_var]])
  
  # Create the violin plot
  p <- ggplot(df, aes_string(x = cat_var, y = continuous_var)) +
    geom_point(fill = "lightblue", color = "darkblue") +
    stat_summary(fun = mean, geom = "point", shape = 20, size = 3, color = "red") +
    labs(title = paste("Violin Plot of", continuous_var, "by", cat_var),
         x = cat_var,
         y = continuous_var) +
    theme_minimal()
  
  # # Print each plot
  print(p)
  
  # Perform ANOVA
  anova_result <- aov(as.formula(paste(continuous_var, "~", cat_var)), data = df)
  
  # Extract F-value and p-value from ANOVA summary
  anova_summary <- summary(anova_result)[[1]]
  f_value <- anova_summary["F value"][1, 1]
  p_value <- anova_summary["Pr(>F)"][1, 1]
  
  # Calculate Eta Squared
  eta_sq <- eta_squared(anova_result, partial = FALSE)
  eta_sq_value <- eta_sq$Eta2[1]
  
  # Append results to the anova_results data frame
  anova_results <- rbind(anova_results, data.frame(
    Categorical_Variable = cat_var,
    F_Value = f_value,
    P_Value = p_value,
    Eta_Squared = eta_sq_value
  ))
}

# Display the final table with all ANOVA summaries and effect sizes
View(anova_results)
```

```r
categorical_vars <- c("age", "race.ethnicity", "chronic.risk.factor..asthma..diabetes..hbp", "disability.status", "covered.by.medicare", "type.of.current.health.insurance..under.65")

# Initialize an empty matrix to store Cramér's V values
correlation_matrix <- matrix(NA, nrow = length(categorical_vars), ncol = length(categorical_vars),
                             dimnames = list(categorical_vars, categorical_vars))

# Loop through each pair of variables to calculate Cramér's V
for (i in 1:length(categorical_vars)) {
  for (j in i:length(categorical_vars)) {
    if (i == j) {
      # Set diagonal to NA or 1 if you prefer
      correlation_matrix[i, j] <- NA
    } else {
      # Create a contingency table for the pair of variables
      contingency_table <- table(df[[categorical_vars[i]]], df[[categorical_vars[j]]])
      
      # Calculate Cramér's V
      cramers_v <- assocstats(contingency_table)$cramer
      correlation_matrix[i, j] <- cramers_v
      correlation_matrix[j, i] <- cramers_v  # Symmetric matrix
    }
  }
}

# Print the correlation matrix 
View(correlation_matrix)
```

```r
# Create folds and write out folds 
set.seed(1)
df$fold <- sample(rep(1:5, length.out = nrow(df)))
```

```r
# Calculate cross validation error on regression
cross_validation_error <- function(model_function, predict_function, data, target_var, n_folds = 5) {
  # Initialize a vector to store the error for each fold
  mse_values <- numeric(n_folds)
  
  # Loop over each fold
  for (i in 1:n_folds) {
    # Split data into training and test sets
    test_data <- data[data$fold == i, ]
    train_data <- data[data$fold != i, ]
    
    # Fit the model on the training data
    model <- model_function(train_data)
    
    # Get predictions on the test data
    predictions <- predict_function(model, test_data)
    
    # Calculate Mean Squared Error for the current fold
    actuals <- test_data[[target_var]]
    mse_values[i] <- mean((actuals - predictions)^2)
  }
  
  # Return the average MSE across all folds
  mean(mse_values)
}
```

```r
# 2. Quantile Regression using covariates that show correlation
set.seed(1)

# fit to all data
formula <- as.formula("X..of.times.saw.a.doctor.in.past.year ~ 
                      1 + race.ethnicity + + disability.status + chronic.risk.factor..asthma..diabetes..hbp + gender + 
                      covered.by.medicare + serious.psychological.distress..kessler.score")
model <- rq(formula, data = df, tau = 0.68)
print(summary(model))

# calculate cross-validation error
tau_values <- seq(0, 1, by = 0.01)
errors <- c()

# Loop over tau values to find the best tau based on CV error
for (tau in tau_values) {
  model_function <- function(data) {
      rq(formula, data = data, tau = tau)
    }
  cv_error <- cross_validation_error(model_function, predict_function, df, "X..of.times.saw.a.doctor.in.past.year", 5)
  errors <- c(errors, cv_error)
}

# Find the tau with the minimum cross-validation error
optimal_tau <- tau_values[which.min(errors)]
cat("Optimal tau:", optimal_tau, "with minimum CV error:", min(errors), "\n")
```

```r
# 3. Lasso regression
set.seed(1)
options(warn=-1)
# Custom predict function for lasso regression with glmnet, ensuring column alignment
predict_lasso <- function(model_info, newdata) {
  # Extract model and feature names
  model <- model_info$model
  feature_names <- model_info$feature_names
  
  # Ensure newdata has the same columns as the training data
  X_new <- model.matrix(~ . - 1, data = newdata[ , setdiff(names(newdata), c(continuous_var, "fold"))])
  
  # Predict using the model at the specific lambda
  as.vector(predict(model, newx = X_new, s = model$lambda))
}
# calculate cross-validation error
lambda_values <- seq(0.01, 1, by = 0.01)
errors <- c()

for (lambda in lambda_values) {
  lasso_model_function <- function(data) {
      # Separate features and target variable
      X <- model.matrix(~ . - 1, data = data[ , setdiff(names(data), c(continuous_var, "fold"))])
      y <- data[[continuous_var]]
      
      # Fit lasso regression model using glmnet
      model <- glmnet(X, y, alpha = 1, lambda = lambda)
      
      # Return the model object and feature names for alignment
      return(list(model = model, feature_names = colnames(X)))
    }
  cv_error <- cross_validation_error(lasso_model_function, predict_lasso, df, "X..of.times.saw.a.doctor.in.past.year", 5)
  errors <- c(errors, cv_error)
}

# Find the tau with the minimum cross-validation error
optimal_lambda <- tau_values[which.min(errors)]
cat("Optimal Lambda:", optimal_lambda, "with minimum CV error:", min(errors), "\n")

X <- model.matrix(~ . - 1, data = df[ , setdiff(names(df), c(continuous_var, "fold"))])
Y <- df[[continuous_var]]
model <- glmnet(X, Y, alpha = 1, lambda = optimal_lambda)
predictions <- predict(model, newx = X, s = optimal_lambda)
lasso_coefs <- coef(model, s = optimal_lambda)

# Print out the coefficients of covariates
print(lasso_coefs)
# Calculate R^2
ss_res <- sum((Y - predictions)^2)
ss_tot <- sum((Y - mean(Y))^2)
r_squared <- 1 - (ss_res / ss_tot)

cat("R-squared:", r_squared, "\n")
# Calculate Mean Squared Error for the current fold
actuals <- df[[continuous_var]]
mse <- mean((actuals - predictions)^2)
cat("\n", "Error on train set: ", mse)
```
